### Duplicates check

```python
# checking duplicates
key_cols = [
    'Gender',
    'Age',
    'Customer Type',
    'Loyalty Program',
    'Type of Travel',
    'Class',
    'Flight Distance',
    'Departure Delay in Minutes',
    'Arrival Delay in Minutes'
]

dupl_key = int(df.duplicated(subset=key_cols).sum())
print(f'there are {dupl_key} duplicates by key colums: \n {key_cols}')
```

***Duplicate checks and repeated profiles***

To check for potential duplicates, I looked at rows that share the same combination of key characteristics: gender, age, customer type, type of travel, class, flight distance, and departure/arrival delays. This produced about 4.6K rows that are flagged as duplicates *by this key*, meaning that there are many passengers who fall into identical profile combinations.

This does **not** mean that rows are literally copied. In this dataset, most variables are discrete and low-cardinality (age in whole years, service ratings from 0–5, delays in minutes, only a few travel classes and customer types). With thousands of observations, it is expected that many passengers share exactly the same profile across these fields. In other words, the duplicate check is mostly detecting **repeated passenger types**, not data entry errors.

I also inspected example rows from these groups and found that other fields (such as detailed service ratings or flight distances in the wider dataset) vary across records, and there is no evidence of inconsistent labels for clearly identical entries. Based on this, I treat these cases as normal repeated profiles rather than true duplicate records, and I do **not** drop them at this stage.

The **Baggage handling** column is the only service-related variable recorded exclusively on a 1–5 scale, with no zero values observed. All other service rating variables use a 0–5 scale. This contrast suggests that a value of 0 in the other service columns likely represents a “not applicable” or “not used” case rather than a poor service rating. Based on this interpretation, zero values in service ratings are treated as non-applicable observations rather than negative evaluations.

### Note on dataset complexity

Although the Airline Passenger Satisfaction dataset initially appears to be a straightforward survey dataset, deeper exploration reveals a richer structure. The data combines passenger demographics, travel context, operational performance (such as delays and flight distance), and multiple service quality measures with different measurement conventions. Some service features are always applicable, while others include explicit non-applicable values, and there are edge cases such as very young passengers labeled as loyal customers. Together, these characteristics indicate that the dataset supports more nuanced analysis than a simple satisfaction prediction task.
